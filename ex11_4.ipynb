- Tạo 1000 dữ liệu kiểu int ngẫu nhiêu gắn với 1000 ngày từ 09/02/2010.
- Dùng kiểu pd.Series.

- Tính cumsum() và vẽ đồ thị.

```python
import numpy as np
```

```python
import pandas as pd
```

```python
data = pd.Series(data = np.random.randint(100, size = 1000), index = pd.date_range('2010/02/09', periods=1000))
```

```python
data = np.cumsum(data)
data.plot()
```

Tìm ra 10 tỉnh có diện tích lớn nhất Việt Nam năm 2015.
Vẽ đồ thị cột (bar) diện tích của 10 tỉnh đó.

```python
import pandas as pd
```

```python
path = '/mnt/dataS/pymi/pyfml/exercises/dientich.csv'
```

```python
data = pd.read_csv(path, index_col=0, sep=';', skiprows=[0, 3], usecols=[0, 5])
data
```

```python
data = data.nlargest(10, columns='2015')
data
```

```python
data.plot.bar()
```
